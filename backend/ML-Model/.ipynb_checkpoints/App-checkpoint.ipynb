```python
from joblib import load
import numpy as np
from flask import Flask, request, jsonify, send_from_directory
import base64
from PIL import Image
from io import BytesIO
import easyocr,os,json
# Load the model from the file

app = Flask(__name__);
@app.route('/')
def home():
    return jsonify({"hello" : "word"})

@app.route('/getPrediction', methods=['POST'])
def getPred():
    df = pd.read_csv('cleaned_data.csv');
    array = np.zeros((1,657))
    inputValues = request.json['symptoms'];
    finalValues = inputValues.lower();
    print(finalValues);
    stringArray = finalValues.split(',');
    for i in range(df.columns.size):
    #     if df.columns[i].lower() == "persistent sore throat":
    #         array[0][i-1] = 1;
    #     elif df.columns[i].lower() == "diarrhea":
    #         array[0][i-1] = 1;
    #     elif df.columns[i].lower() == "sore throat":
    #         array[0][i-1] = 1;
        if df.columns[i].lower() in stringArray:
            array[0][i-1] = 1;
            print(i)
    print(array)
    loaded_model = load('decision_tree_model.pkl')
    # Make predictions
    predictions = loaded_model.predict(array);
    print(predictions[0]);
    return jsonify({"prediction" : predictions[0]});


UPLOAD_FOLDER = 'processed_images'
if not os.path.exists(UPLOAD_FOLDER):
    os.makedirs(UPLOAD_FOLDER)

@app.route('/process', methods=['POST'])
def process_image():
    image_base64 = request.json.get('image')

    if not image_base64:
        return jsonify({'error': 'No image data provided'}), 400

    try:
        # Decode the base64 encoded image
        image_data = base64.b64decode(image_base64)
        
        # Open the image using PIL
        image = Image.open(BytesIO(image_data))
        
        # Generate a unique filename
        filename = f"{os.urandom(24).hex()}.png"

        # Save the processed image to the processed_images folder
        image.save(os.path.join(UPLOAD_FOLDER, filename))
        
        reader = easyocr.Reader(['en'])

        dir_path = r"./processed_images"
        image_path=dir_path+"/"+os.listdir(dir_path)[0]
        result = reader.readtext(image_path)
        flag = 0 
        l=[]
        for i in range(len(result)):
            if result[i][1] == "List of Prescribed Medications":
                flag = flag + 1
            if flag > 0 and flag < 40:
                flag= flag+1
                if result[i][1] == "Physician Signature":
                    break
                l.append(result[i][1])
        l2=[]
        for i in range(len(l)-2):
            for j in l[i].split(" - "):
                l2.append(j)
        med_l=[]
        dos_l=[]
        x=[]
        flag=0
        for i in range(len(l2)):
            # print(l2[i])/
            if i==1 or i==3 or i==6 or i==8 or i==11 or i==13 or i==16 or i==18 or i==21 or i==23 or i==26 or i==28:
                x.append(l2[i])
        for i in range(len(x)):
            if i%2==0:
                med_l.append(x[i])
            else:
                dos_l.append(x[i])
        json_data=[]
        for i,j in zip(med_l,dos_l):
            dict={}
            dict["medicine"]=i
            dict["dosage"]=j
            json_data.append(dict)
        print(json_data)
        
        return jsonify({"message": json_data})
    except Exception as e:
        return jsonify({'error': f'Error processing image: {str(e)}'}), 500

if __name__ == '__main__':
    app.run(port=8000)
```
